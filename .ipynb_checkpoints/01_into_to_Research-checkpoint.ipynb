## Sampling
- draw data from a normal distribution 

```python
# Sample 100 points with a mean of 0 and a standard deviation (std) of 1 -> this is a standard normal distribution
#and you draw randomly from this distrubution 
x = np.random.normal(0,1,100)
#0 is the center of the distribution, where is most data coming from 
#1 is how wide is the distriution, how likely is it to has extrema 
#100 how large is the sample
```

## Adding Axis Labels


```python
#Two x'es, both sampled from random data
x = np.random.normal(0,1,100)
x2 = np.random.normal(0,1,100)

plt.plot(x); #see semicolon here to avoid function output on screen
plt.plot(x2);

#Labelling the Axes
plt.xlabel('Time') #The data we generated is unitless but just for practise don't forget units in general
plt.ylabel('Returns')
#Legend takes in an array and the elements in the array are the different series you plotted
plt.legend(['X', 'X2']);
```

## Generating Statistics 
- Let's use Numpy to take some simple statistics
- As we were generating the data we were passing a zero mean and a 1 standard deviation so this data is quite close
- as Sample Size grows larger (let's say from 100 to 10000) we'll observe the mean and standard deviation will move much closer to what we chose for the values, they'll converge as we have more samples


```python
np.mean(x)
```

```python
#Standard Deviation
np.std(x)
```
